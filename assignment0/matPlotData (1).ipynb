```python
x = np.arange(0, 101, 1)
y = 0.5 * x**2 + 2 * x + 1

plt.plot(x, y, color = 'orange', label = r'$y = 0.5x^2 + 2x + 1$')


plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title('Quadratic Function Plot')
plt.legend()

plt.grid(True)

plt.savefig('quadratic_function.pdf', format = 'pdf')

plt.show()
```
