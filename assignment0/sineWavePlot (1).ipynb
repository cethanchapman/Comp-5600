```python
x = np.arange(0, 100, 0.1)

y = np.sin(x)

plt.xlabel("x values\n0 to 100")
plt.ylabel("sin(x)")
plt.title("Sine Wave: y = sin(x)")

plt.grid(True)
plt.plot(x, y)

current_directory = os.getcwd()
print(f"Current working directory: {current_directory}")

file_path = os.path.join(current_directory, 'sine_wave.png')
plt.savefig(file_path)

plt.show()
```
